```python
import pandas as pd
from sklearn.metrics import accuracy_score

df = pd.read_csv('train.csv', index_col=0)
y = df['Survived']

ypred = []
for i, row in df.iterrows():
    if row['Sex'] == 'female':
        ypred.append(1)
    else:
         
        ypred.append(0)

print(accuracy_score(y, ypred))
```
