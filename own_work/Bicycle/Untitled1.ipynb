```python
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
df = sns.load_dataset('flights')
```

```python
df['stringDate']=df.apply(lambda row: str(row['year'])+' ' + row['month'],axis=1)  
#Converts two columns to a string and combines to new column
```

```python
df['time']=pd.to_datetime(df['stringDate'])
```

```python
df['x']=range(144)
```

```python
X=df[['time',]]
```

```python
y=df['passengers']
```

```python
df.dtypes
```

```python
m = LinearRegression()
```

```python
m.fit(X,y)
```

```python
df['ypred'] = m.predict(df[['x']])
```

```python
df[['passengers','ypred']].plot()
```
