```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
import tensorflow as tf
import numpy as np
from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt


```

```python

K.clear_session()
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()

X = X_train.reshape(len(X_train),784)
y = tf.keras.utils.to_categorical(y_train, num_classes=10)

```

```python
model = Sequential([
    Dense(930, input_shape=(784,)),
    Activation('sigmoid'),
    Dense(250),
    Activation('relu'),
    Dense(10),
    Activation('softmax'),
])

```

```python
model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X, y, epochs=5,  verbose=1)
```

```python
model.summary()
```

```python
history = model.fit(X, y, validation_split=0.33, epochs=5, batch_size=10, verbose=0)
```

```python

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
# summarize history for accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
score = model.evaluate(X, y, batch_size=10000)
print(score)

print(model.predict(X))
```
